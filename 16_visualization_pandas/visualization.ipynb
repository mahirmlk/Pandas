# Visualization with Pandas

Using Pandas built-in plotting for quick and effective data visualization.


```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'date': pd.date_range('2023-01-01', periods=10),
    'sales': [100, 120, 130, 125, 140, 160, 155, 170, 180, 190],
    'profit': [20, 25, 30, 28, 32, 40, 38, 42, 45, 50]
}).set_index('date')

df
```

## Built-in plotting

```python
df.plot()
```

## Line plots

```python
df['sales'].plot()
```

```python
df[['sales', 'profit']].plot()
```

## Bar plots

```python
df.sum().plot(kind='bar')
```

## Histogram

```python
df['sales'].plot(kind='hist', bins=5)
```

## Box plots

```python
df[['sales', 'profit']].plot(kind='box')
```

## Time series plots

```python
df['sales'].plot()
```

## Customization basics

```python
ax = df['sales'].plot(title='Sales Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Sales')
```

## When to switch

Switch to Matplotlib or Seaborn when:
- You need fine-grained styling control
- You are building complex multi-plot layouts
- Statistical plots or themes are required
