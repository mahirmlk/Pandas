# Data Inspection & Exploration

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai'],
    'sales': [100, 150, 120, 90, 160],
    'discount': [5, 10, np.nan, 7, 10]
})

df.head()
```

```python
df.tail()
```

```python
df.sample(2, random_state=1)
```

## Structure

```python
df.info()
```

## Summary statistics

```python
df.describe()
```

## Data types

```python
df.dtypes
```

## Value counts

```python
df['city'].value_counts()
```

## Unique values

```python
df['city'].unique()
```

```python
df['city'].nunique()
```

## Memory usage

```python
df.memory_usage(deep=True)
```


- Start with `head()`, `info()`, and `describe()`
- Verify `dtypes` before analysis
- Use `value_counts()` for categorical columns
- Monitor memory early for scalability
