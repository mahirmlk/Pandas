# Indexing, Selection & Filtering

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'city': ['Delhi', 'Mumbai', 'Delhi', 'Chennai', 'Mumbai'],
    'year': [2021, 2021, 2022, 2022, 2023],
    'sales': [100, 150, 120, 90, 160],
    'profit': [20, 35, 25, 10, 40]
})

df
```

## Column selection

```python
df['sales']
```

```python
df[['city', 'profit']]
```

```python
df.sales
```

```python
df[['sales']].head()
```

## Row selection

```python
df.loc[0]
```

```python
df.loc[1:3]
```

```python
df.iloc[0]
```

```python
df.iloc[1:4]
```

```python
df[df['sales'] > 120]
```

```python
df[(df['sales'] > 100) & (df['city'] == 'Mumbai')]
```

```python
df.query("sales > 120")
```

```python
df.query("city == 'Delhi' and year == 2022")
```

## Index operations

```python
df_indexed = df.set_index('city')
df_indexed
```

```python
df_indexed.loc['Mumbai']
```

```python
df_reset = df_indexed.reset_index()
df_reset
```

```python
df_multi = df.set_index(['city', 'year'])
df_multi
```

```python
df_multi.loc[('Delhi', 2022)]
```

```python
df_multi.loc['Mumbai']
```
