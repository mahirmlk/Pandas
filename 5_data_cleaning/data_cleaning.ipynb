# Data Cleaning

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Bob', 'Charlie', None],
    'age': [25, np.nan, 30, 35, 40],
    'salary': ['50000', '60000', '60000', None, '70000'],
    'join_date': ['2021-01-01', '2021-02-15', '2021-02-15', None, '2022-03-10']
})

df
```

## Missing data

```python
df.isnull()
```

```python
df.notnull()
```

```python
df.dropna()
```

```python
df.dropna(subset=['salary'])
```

```python
df.fillna(0)
```

```python
df.fillna({'age': df['age'].mean(), 'salary': '0'})
```

```python
df.ffill()
```

```python
df.bfill()
```

## Duplicates

```python
df.duplicated()
```

```python
df.duplicated(subset=['name', 'join_date'])
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(subset=['name'], keep='first')
```

```python
df.drop_duplicates(subset=['name'], keep='last')
```

## Data types

```python
df.dtypes
```

```python
df['salary'] = df['salary'].astype('float')
df.dtypes
```

```python
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')
df['salary']
```

```python
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')
df['join_date']
```

```python
df['name'] = df['name'].astype('category')
df.dtypes
```
